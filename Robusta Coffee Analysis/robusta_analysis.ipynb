```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter
```

## COT Report

Source: https://www.ice.com/report/122

```python
cot_hist_dfs = []
for year in range(2020, 2025):
    filename = f"COTHist{year}.csv"
    if os.path.exists(filename):
        df = pd.read_csv(filename)
        cot_hist_dfs.append(df)


cot_hist = pd.concat(cot_hist_dfs, ignore_index=True)


cot_hist.tail()
```

```python
cot_hist.head()
```

```python
pd.set_option("display.max_columns", None)
```

```python
cot_hist["Market_and_Exchange_Names"].info()
```

```python
cot_hist_robusta = cot_hist[cot_hist["Market_and_Exchange_Names"].str.contains("Robusta")].copy()
cot_hist_robusta = cot_hist_robusta.dropna(axis=1, how="all")

cot_hist_robusta.head()
```

```python
cot_hist_robusta.fillna(0, inplace=True)
```

```python
columns_to_drop = [col for col in cot_hist_robusta.columns if any(substring in col for substring in ["Old", "Traders", "Pct"])]
cot_hist_robusta.drop(columns=columns_to_drop, inplace=True)

cot_hist_robusta.head()
```

```python
opt_cot_hist_robusta = cot_hist_robusta[cot_hist_robusta["FutOnly_or_Combined"] == "Combined"].copy()
```

```python
opt_cot_hist_robusta.tail()
```

```python
opt_cot_hist_robusta["Commercial_Long"] = opt_cot_hist_robusta["Prod_Merc_Positions_Long_All"] + opt_cot_hist_robusta["Swap_Positions_Long_All"]
opt_cot_hist_robusta["Commercial_Short"] = opt_cot_hist_robusta["Prod_Merc_Positions_Short_All"] + opt_cot_hist_robusta["Swap_Positions_Short_All"]

opt_cot_hist_robusta["Non_Comm_Long"] = opt_cot_hist_robusta["M_Money_Positions_Long_All"] + opt_cot_hist_robusta["Other_Rept_Positions_Long_All"]
opt_cot_hist_robusta["Non_Comm_Short"] = opt_cot_hist_robusta["M_Money_Positions_Short_All"] + opt_cot_hist_robusta["Other_Rept_Positions_Short_All"]
```

```python
opt_cot_hist_robusta.tail()
```

```python
opt_cot_hist_robusta = opt_cot_hist_robusta[["Market_and_Exchange_Names", "As_of_Date_Form_MM/DD/YYYY", "Open_Interest_All", "Commercial_Long", "Commercial_Short",\
                                             "Non_Comm_Long", "Non_Comm_Short"]].copy()
```

```python
opt_cot_hist_robusta.tail()
```

```python
opt_cot_hist_robusta["Commercial_Net"] = opt_cot_hist_robusta["Commercial_Long"] - opt_cot_hist_robusta["Commercial_Short"]

opt_cot_hist_robusta["Non_Comm_Net"] = opt_cot_hist_robusta["Non_Comm_Long"] - opt_cot_hist_robusta["Non_Comm_Short"]

opt_cot_hist_robusta.tail()
```

```python
commercial_long = opt_cot_hist_robusta["Commercial_Long"]
commercial_short = opt_cot_hist_robusta["Commercial_Short"]
commercial_net = opt_cot_hist_robusta["Commercial_Net"]
dates = pd.to_datetime(opt_cot_hist_robusta["As_of_Date_Form_MM/DD/YYYY"])


min_date = pd.Timestamp(year=dates.min().year, month=1, day=1)

fig, ax = plt.subplots(figsize=(10, 6))


ax.fill_between(dates, commercial_long, color="#176640", alpha=0.8, label="Commercial Long")

ax.fill_between(dates, -commercial_short, color="#0b5394", alpha=0.8, label="Commercial Short")

ax.plot(dates, commercial_net, color="black", linestyle="--", label="Commercial Net")


ax.set_ylabel("Volume")
ax.set_title("Robusta - Commercial Positions", fontweight="bold")
ax.legend()

ax.set_xlim(min_date, dates.max())

ax.set_yticks(range(-150000, 180000, 30000))


tick_dates = pd.date_range(start=min_date, end=dates.max(), freq="3M")
tick_labels = [date.strftime("%b\n%y") for date in tick_dates]
ax.set_xticks(tick_dates)
ax.set_xticklabels(tick_labels)

ax.grid(True, linestyle="--", alpha=0.2, color="black")


plt.tight_layout()
plt.show()
```

```python
non_comm_long = opt_cot_hist_robusta["Non_Comm_Long"]
non_comm_short = opt_cot_hist_robusta["Non_Comm_Short"]
non_comm_net = opt_cot_hist_robusta["Non_Comm_Net"]
dates = pd.to_datetime(opt_cot_hist_robusta["As_of_Date_Form_MM/DD/YYYY"])

min_date = pd.Timestamp(year=dates.min().year, month=1, day=1)

fig, ax = plt.subplots(figsize=(10, 6))


ax.fill_between(dates, non_comm_long, color="#176640", alpha=0.8, label="Non-Comm Long")

ax.fill_between(dates, -non_comm_short, color="#0b5394", alpha=0.8, label="Non-Comm Short")

ax.plot(dates, non_comm_net, color="black", linestyle="--", label="Non-Comm Net")


ax.set_ylabel("Volume")
ax.set_title("Robusta Non-Comm Positions", fontweight="bold")
ax.legend()

ax.set_xlim(min_date, dates.max())

ax.set_yticks(range(-70000, 60000, 20000))



tick_dates = pd.date_range(start=min_date, end=dates.max(), freq="3M")
tick_labels = [date.strftime("%b\n%y") for date in tick_dates]
ax.set_xticks(tick_dates)
ax.set_xticklabels(tick_labels)

ax.grid(True, linestyle="--", alpha=0.2, color="black")


plt.tight_layout()
plt.show()
```

```python
robusta_price["Price"] = robusta_price["Price"].str.replace(",", "").astype(float)
robusta_price["Date"] = pd.to_datetime(robusta_price["Date"], format="%m/%d/%Y")
```

## Production vs Consumption & Stock to Use

### Production vs Consumption

```python
usda_data = pd.read_csv("psd_coffee.csv")
```

```python
usda_data.info()
```

```python
usda_data.head()
```

```python
usda_data = usda_data[usda_data["Market_Year"] >= 2010]
```

```python
usda_data["Attribute_Description"].value_counts()
```

```python
usda_consumption = usda_data[usda_data["Attribute_Description"] == "Domestic Consumption"]
usda_consumption = usda_consumption.groupby("Market_Year")["Value"].sum().reset_index()

usda_consumption.tail()
```

```python
usda_production = usda_data[usda_data["Attribute_Description"] == "Production"]
usda_production = usda_production.groupby("Market_Year")["Value"].sum().reset_index()

usda_production.tail()
```

```python
usda_p_vs_c = pd.merge(usda_production, usda_consumption, how = "outer", on="Market_Year")

usda_p_vs_c.tail()
```

### Stock to use

```python
usda_ending_stocks = usda_data[usda_data["Attribute_Description"] == "Ending Stocks"]
usda_ending_stocks = usda_ending_stocks.groupby("Market_Year")["Value"].sum().reset_index()

usda_ending_stocks.tail()
```

```python
usda_stock_to_use = pd.merge(usda_consumption, usda_ending_stocks, how="outer", on="Market_Year")

usda_stock_to_use.tail()
```

```python
usda_stock_to_use["Stock_to_Use"] = usda_stock_to_use["Value_y"] / usda_stock_to_use["Value_x"]

usda_stock_to_use
```

### Graph

```python
usda_stock_to_use = usda_stock_to_use[["Market_Year", "Stock_to_Use"]]
```

```python
usda_pc_stock = pd.merge(usda_p_vs_c, usda_stock_to_use, how="outer", on="Market_Year")
usda_pc_stock.rename(columns={"Value_x": "Production", "Value_y": "Consumption"}, inplace=True)
```

```python
usda_pc_stock
```

```python
fig, ax1 = plt.subplots(figsize=(10, 6))

ax1.bar(usda_pc_stock["Market_Year"] - 0.2, usda_pc_stock["Production"], width=0.35, color="#0b5394", alpha=0.8, label="Production")
ax1.bar(usda_pc_stock["Market_Year"] + 0.2, usda_pc_stock["Consumption"], width=0.35, color="#93c47d", alpha=0.8, label="Consumption")

ax1.set_xlabel("Market Year")
ax1.set_ylabel("Volume")
ax1.set_title("Production and Consumption vs Stock-to-Use", fontweight="bold")

ax1.set_yticks(range(0, 200001, 25000))

ax1.legend(loc="upper left")

# Add gridlines
ax1.grid(True, linestyle="--", linewidth=0.5, color="black", alpha=0.2)

ax2 = ax1.twinx()
ax2.plot(usda_pc_stock["Market_Year"], usda_pc_stock["Stock_to_Use"], color="#444444", label="Stock-to-Use", alpha=0.7, marker="o")
ax2.set_ylabel("Stock-to-Use (%)")
ax2.legend(loc="upper right")

formatter = FuncFormatter(lambda x, _: "{:.2f}%".format(x * 100))
ax2.yaxis.set_major_formatter(formatter)

ax1.set_xticks(usda_pc_stock["Market_Year"])


plt.tight_layout()
plt.show()
```

## Vietnam Exports

```python
usda_data["Attribute_Description"].value_counts()
```

```python
vietnam_data = usda_data[usda_data["Country_Name"] == "Vietnam"]

vietnam_data.head()
```

```python
vietnam_exports = vietnam_data[vietnam_data["Attribute_Description"] == "Exports"]
vietnam_exports = vietnam_exports[["Market_Year", "Value"]]
vietnam_exports.rename(columns={"Value": "Exports"}, inplace=True)

vietnam_exports.head(20)
```

```python
vietnam_consumption = vietnam_data[vietnam_data["Attribute_Description"] == "Domestic Consumption"]
vietnam_consumption = vietnam_consumption[["Market_Year", "Value"]]
vietnam_consumption.rename(columns={"Value": "Consumption"}, inplace=True)

vietnam_consumption.head(20)
```

```python
vietnam_ending_stocks = vietnam_data[vietnam_data["Attribute_Description"] == "Ending Stocks"]
vietnam_ending_stocks = vietnam_ending_stocks[["Market_Year", "Value"]]
vietnam_ending_stocks.rename(columns={"Value": "Ending Stocks"}, inplace=True)

vietnam_ending_stocks.head(20)
```

```python
vietnam_production = vietnam_data[vietnam_data["Attribute_Description"] == "Production"]
vietnam_production = vietnam_production[["Market_Year", "Value"]]
vietnam_production.rename(columns={"Value": "Production"}, inplace=True)

vietnam_production.head(20)
```

```python
vietnam_graph_data = pd.merge(pd.merge(pd.merge(vietnam_consumption, vietnam_exports, on="Market_Year"),\
                                   vietnam_ending_stocks, on="Market_Year"), vietnam_production, on="Market_Year")

vietnam_graph_data["Supply"] = vietnam_graph_data["Ending Stocks"] + vietnam_graph_data["Production"]
vietnam_graph_data["Demand"] = vietnam_graph_data["Consumption"] + vietnam_graph_data["Exports"]
vietnam_graph_data["Net"] = vietnam_graph_data["Supply"] - vietnam_graph_data["Demand"]


vietnam_graph_data = vietnam_graph_data[["Market_Year", "Supply", "Demand", "Net"]]

vietnam_graph_data
```

```python
fig, ax1 = plt.subplots(figsize=(10, 6))

ax1.plot(vietnam_graph_data["Market_Year"], vietnam_graph_data["Supply"], color="#176640", label="Supply", linewidth=3)

ax1.plot(vietnam_graph_data["Market_Year"], vietnam_graph_data["Demand"], color="#0b5394", label="Demand", linewidth=3)


ax1.set_xlabel("Market Year")
ax1.set_ylabel("Supply/Demand", color="black")
ax1.set_title("Vietnam - Supply and Demand", fontweight="bold")
ax1.tick_params(axis="y", labelcolor="black")

ax1.grid(True, linestyle="--", linewidth=0.5, color="gray", alpha=0.5)

ax1.legend(loc="upper left")


ax2 = ax1.twinx()

ax2.bar(vietnam_graph_data["Market_Year"], vietnam_graph_data["Net"], color="#cc0808", label="Net Value")
ax2.legend(loc="upper right")

ax2.set_ylabel("Net Value", color="black")
ax2.tick_params(axis="y", labelcolor="black")


ax1.set_ylim(0, 40000)  
ax2.set_ylim(-5000, 15000) 


ax1.set_xticks([year for year in vietnam_graph_data["Market_Year"] if year != 2024])


# Show plot
plt.tight_layout()
plt.show()
```
